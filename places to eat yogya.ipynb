```python
dict_change = {'Barat':'Western', 'Cepat saji':'Fast food', 'Ayam & bebek':'Chicken & duck', 'Korean':'Korea',
               'Beverages':'Minuman', 'Bakmie':'Noodles', 'Aneka nasi':'Rice', 'Jajanan':'Snacks'}
```

```python
def replace_variasi(variasi):
    for key in dict_change.keys():
        if key in variasi:
            variasi = variasi.replace(key, dict_change[key])
    return variasi
```

```python
df['Variasi Makanan']
```

```python
df.apply(lambda row:replace_variasi(row['Variasi Makanan']), axis=1)
```

```python
df['Variasi Makanan'] = df.apply(lambda row:replace_variasi(row['Variasi Makanan']), axis=1)
```

```python
for i in set(variasi.split(', ')):
    df["VM_{}".format(i)] = df.apply(lambda row: row['Variasi Makanan'].count(i), axis=1)
```

```python
df.head()
```

```python
variasi_cols = []
for i in set(variasi.split(', ')):
    variasi_cols.append("VM_{}".format(i))
```

```python
df[variasi_cols].corr().style.background_gradient(cmap='Blues')
```

```python
corre = {}
corre['col_a'], corre['col_a_sum'], corre['col_b'], corre['col_b_sum'], corre['corr'], corre['col_a_b_sum'] = [], [], [], [], [], []
for i in range(len(variasi_cols)):
  for j in range(i+1, len(variasi_cols)):
    corre['col_a'].append(variasi_cols[i])
    corre['col_a_sum'].append(df[variasi_cols[i]].sum())
    corre['col_b'].append(variasi_cols[j])
    corre['col_b_sum'].append(df[variasi_cols[j]].sum())
    corre['corr'].append(df[variasi_cols[i]].corr(df[variasi_cols[j]]))
    corre['col_a_b_sum'].append(len(df.loc[(df[variasi_cols[i]] == 1) & (df[variasi_cols[j]] == 1)]))
    
df_corre = pd.DataFrame(corre)
df_corre.sort_values('corr', ascending=False).head(10).style.background_gradient(cmap='Blues')
# df_corre
```
